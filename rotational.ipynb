```python
import matplotlib.pyplot as plt 
import pandas as pd 

%matplotlib inline
df=pd.read_excel('./rate.xlsx')
x=df["J"]
y=df["Rate_constant"]
err=df['Error']
estm=df['estimate']
fig = plt.figure(figsize=(7,4), dpi=90, facecolor='W', edgecolor='b',linewidth=10)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,".")
ax.plot(x,estm)
ax.plot(x,y)
ax.errorbar(x,y,yerr=err,fmt=' ')
plt.yscale('log')
ax.set_title("Rotational rate constants for j=32 $Li_2$-rare gas collision")
ax.set_xlabel("$J_i$")
ax.set_ylabel("$K_{if}$")
plt.show()
```
